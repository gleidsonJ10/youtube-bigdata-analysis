# Projeto PySpark - Leitura e Escrita de Dados

## 1. Iniciar Spark

```
from pyspark.sql import SparkSession

spark = SparkSession.builder     .appName("Leitura e Escrita")     .getOrCreate()
```

## 2. Ler arquivo videos-stats.csv

```
df_videos = spark.read.csv("videos-stats.csv", header=True)
df_videos.show(8)
```

## 3. Visualizar esquema

```
df_videos.printSchema()
```

## 4. Ler novamente inferindo esquema

```
df_videos_inf = spark.read.csv("videos-stats.csv", header=True, inferSchema=True)
df_videos_inf.printSchema()
```

## 5. Salvar como parquet

```
df_videos_inf.write.mode("overwrite").parquet("videos-parquet")
```

## 6. Ler parquet

```
df_parquet = spark.read.parquet("videos-parquet")
df_parquet.show(5)
```

## 7. Salvar como tabela

```
df_parquet.write.mode("overwrite").saveAsTable("tb_videos")
```

## 8. Listar tabelas

```
spark.catalog.listTables()
```

## 9. Ler tabela com SQL

```
spark.sql("SELECT * FROM tb_videos").show(5)
```

## 10. Ler comments.csv

```
df_comments = spark.read.csv("comments.csv", header=True, inferSchema=True)
df_comments.printSchema()
```

## 11. Salvar comments como parquet

```
df_comments.write.mode("overwrite").parquet("comments-parquet")
```
